# Transform model_state.csv to state_season.csv

```python
# Read in file
df1 = pd.read_csv('../Resources/data_raw/model_state.csv')
print(df1.shape)
df1.head()
```

```python
# Check data types
df1.dtypes
```

```python
# Check for null values
df1.isnull().sum()
```

```python
# Keep relevant columns
df1 = df1[['STATE_NAME','STUSAB', 'Annual','Fall', 'Spring', 'Summer', 'Winter', 'max_warming_season', 'fips']].copy()

# Rename columns
df1.columns = ['state_name','state_abb', 'annual','fall', 'spring', 'summer', 'winter', 'max_warming_season', 'fips']
print(df1.shape)
df1.head()
```

```python
# Iterate through every row of df1 and build dictionaries that 
# store state names, state abbreviation and fips codes
abb_name_dict = {}
name_abb_dict = {}
fips_dict = {}

for row in df1.itertuples():
    abb_name_dict[row[2]] = row[1]
    name_abb_dict[row[1]]= row[2]
    fips_dict[row[9]] = row[1]
```

```python
# Add additional US states/territories/districts to dictionaries

abb_name_dict['AK'] = 'Alaska'
abb_name_dict['HI'] = 'Hawaii'
abb_name_dict['DC'] = 'District of Columbia'
abb_name_dict['US'] = 'United States'
abb_name_dict['PR'] = 'Puerto Rico'
abb_name_dict['VI'] = 'Virgin Islands'

name_abb_dict['Alaska']= 'AK'
name_abb_dict['Hawaii']= 'HI'
name_abb_dict['District of Columbia']= 'DC'
name_abb_dict['United States']= 'US'
name_abb_dict['Puerto Rico']= 'PR'
name_abb_dict['Virgin Islands']= 'VI'
```

```python
# Drop the state_abb and fips columns
df1.drop(['state_abb', 'fips'], axis = 'columns', inplace = True)

print(df1.shape)

df1.head()
```

```python
# Check final data types 
df1.dtypes
```

# Transform energy sources to state_energy.csv

```python
# Read in data sets
coal_df = pd.read_csv("../Resources/data_raw/use_energy_source_coal.csv")
natgas_df = pd.read_csv("../Resources/data_raw/use_energy_source_natural_gas.csv")
nuclear_df = pd.read_csv("../Resources/data_raw/use_energy_source_nuclear.csv")
petroleum_df = pd.read_csv("../Resources/data_raw/use_energy_source_petroleum.csv")
totalrenew_df = pd.read_csv("../Resources/data_raw/use_energy_source_totalrenew.csv")
biomass_df = pd.read_csv("../Resources/data_raw/use_renew_sector_biomass.csv")
geothermal_df = pd.read_csv("../Resources/data_raw/use_renew_sector_geothermal.csv")
hydropower_df = pd.read_csv("../Resources/data_raw/use_renew_sector_hydropower.csv")
solar_df = pd.read_csv("../Resources/data_raw/use_renew_sector_solar.csv")
wind_df = pd.read_csv("../Resources/data_raw/use_renew_sector_wind.csv")
```

```python
# Store the dataframes in a list
df_list = [coal_df, natgas_df, petroleum_df, nuclear_df, totalrenew_df, 
           biomass_df, geothermal_df, hydropower_df, solar_df, wind_df]

name_list = ['pe_coal', 'pe_naturalgas', 'pe_petroleum', 'pe_nuclear', 'pe_totalrenew',
             're_biomass', 're_geothermal', 're_hydropower','re_solar','re_wind']
```

```python
# Define a function that builds a dataframe of energy consumption sources
def builder1(df_list, name_list): 
    
    # Iterate through each data frame in the dataframe list 
    for i in range(0,len(df_list)):
        
        # Drop the columns of null values
        df_list[i].dropna(axis='columns', inplace = True)
        
        # Melt the dataframe such that the column names would be: State-Year-name_list[i]
        df_list[i] = pd.melt(df_list[i], id_vars=['State'], 
                             value_vars= list(df_list[i].columns[1:]),
                             var_name='Year', value_name = name_list[i])
       
        # Merge the melted dataframe to newdf 
        if i < 1:
            newdf = df_list[i]
        else:
            newdf = newdf.merge(df_list[i], how = 'inner', 
                     left_on = ['State', 'Year'], right_on = ['State', 'Year'])
    return newdf
```

```python
# Build df8
df8 = builder1(df_list, name_list)
df8.head()
```

```python
# Rename colmuns 
df8.columns = ['state_abb', 'year', 'pe_coal', 'pe_naturalgas', 'pe_petroleum',
       'pe_nuclear', 'pe_totalrenew', 're_biomass', 're_geothermal',
       're_hydropower', 're_solar', 're_wind']

df8.head()
```

```python
# Add a state name column
df8['state_name'] = df8['state_abb'].apply(lambda x: abb_name_dict[x])
df8.head()
```

```python
# Rearrange columns and keep only relevant ones
df8 = df8[['state_name', 'year', 'pe_coal', 'pe_naturalgas', 'pe_petroleum',
       'pe_nuclear', 'pe_totalrenew', 're_biomass', 're_geothermal',
       're_hydropower', 're_solar', 're_wind']].copy()

print(df8.shape)
df8.head()
```

```python
# Change the year column to int data type
df8['year'] = df8['year'].astype(int)
```

```python
# Check final data types
df8.dtypes
```
